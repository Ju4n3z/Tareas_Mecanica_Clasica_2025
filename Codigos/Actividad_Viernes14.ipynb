# Cuarto Punto

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir constantes
h = 1  # metro
m = 1  # kilogramo
g = 9.8  # m/s^2

# Calcular el tiempo caracterstico
T_critico = np.sqrt(2 * h / g)

# Rango de valores de T
T = np.linspace(0.1, 2, 100)  # Evitamos T=0 para evitar divisiones por cero

# Definicin de las funciones S1, S2 y S3
S1 = (1/2) * m * (h**2 / T) - (1/2) * m * g * h * T
S2 = (2/3) * m * (h**2 / T) - (2/3) * m * g * h * T
S3 = (9/10) * m * (h**2 / T) - (3/4) * m * g * h * T

# Evaluar las funciones en T_critico
S1_critico = (1/2) * m * (h**2 / T_critico) - (1/2) * m * g * h * T_critico
S2_critico = (2/3) * m * (h**2 / T_critico) - (2/3) * m * g * h * T_critico
S3_critico = (9/10) * m * (h**2 / T_critico) - (3/4) * m * g * h * T_critico

# Graficar las funciones
plt.figure(figsize=(8, 6))
plt.plot(T, S1, label=r'$S_1$ (Caso 1)', linestyle='-', color='b')
plt.plot(T, S2, label=r'$S_2$ (Caso 2)', linestyle='--', color='r')
plt.plot(T, S3, label=r'$S_3$ (Caso 3)', linestyle='-.', color='g')

# Graficar lneas horizontales en S(T_critico)
plt.axhline(S1_critico, color='b', linestyle=':', alpha=0.7)
plt.axhline(S2_critico, color='r', linestyle=':', alpha=0.7)
plt.axhline(S3_critico, color='g', linestyle=':', alpha=0.7)

# Marcar el punto en T_critico
plt.scatter([T_critico] * 3, [S1_critico, S2_critico, S3_critico], color=['b', 'r', 'g'], zorder=3)

# Configuracin de la grfica
plt.xlabel("Tiempo (T) [s]")
plt.ylabel("Accin (S)")
plt.title("Comparacin de las acciones para los tres casos")
plt.legend()
plt.grid(True)

# Mostrar la grfica
plt.show()

print(f"El valor de S1 en T crtico es: {S1_critico}")
print(f"El valor de S2 en T crtico es: {S2_critico}")
print(f"El valor de S3 en T crtico es: {S3_critico}")
```
